```python
import os
import glob
import pandas as pd
os.chdir('./data/test')
```

```python
extension = 'xlsx'
all_filenames = [i for i in glob.glob('*.{}'.format(extension))]
```

```python
#combine all files in the list
combined_csv = pd.concat([pd.read_excel(f,index_col=1, header = 1) for f in all_filenames ])
#export to csv
combined_csv.to_csv( "combined_csv.csv", index=False, encoding='utf-8-sig')
```
